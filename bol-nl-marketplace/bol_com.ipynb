```python
# load libraries
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
from google.colab import files
```

```python
result = list()
# request data from url
for x in range(500):
  url = 'https://www.bol.com/nl/nl/l/boeken/8299/?page='+str(x)+'&_c=xhr&bltgc=teNMEjfzFsXkVRm17xxMMw'
  res = requests.get(url)
  data = BeautifulSoup(res.text)
  
  links = data.find_all('a', class_='product-title')
  for l in range(len(links)):
    url = 'https://www.bol.com' + links[l].get('href')
    result.append(url)
    print(url)

result
```

```python
len(result)
```

```python
product_detail = {}
for x in range(len(result)):
  res = requests.get(result[x])
  data = BeautifulSoup(res.text)

  # get product information
  title = data.select('span[data-test=title]')[0].get_text()
  author = data.select('a[data-role=AUTHOR]')
  if len(author) > 0:
    author = author[0].get_text().strip()
  else:
    author = "No data"
  price = data.select('span[data-test=price]')[0].get_text().strip().replace('\n  ', ',')
  product_rating = data.select('div[data-test=rating-suffix]')
  if len(product_rating) > 0:
    product_rating = product_rating[0].get_text()
  else:
    product_rating = "No reviews"

  # compile product detail
  product_detail[x] = {
      # "url": url
      "title": title,
      "author": author,
      "price": price,
      "product_rating": product_rating
  }

  print(product_detail[x])
  
```

```python
headers = [ "title", "author", "price", "product_rating"]
df = pd.DataFrame.from_dict(product_detail, orient='index')
df.to_excel("product_detail.xlsx")
files.download('product_detail.xlsx')
```
