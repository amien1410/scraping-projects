```python
# load libraries
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
from google.colab import files
```

```python
result = list()
num = 1
# request data from url
for x in range(40):
  url = 'https://celulares.mercadolivre.com.br/_Desde_'+str(num)+'_NoIndex_True'
  res = requests.get(url)
  data = BeautifulSoup(res.text)
  
  links = data.find_all('a', class_='ui-search-item__group__element ui-search-link')
  for l in range(len(links)):
    url = links[l].get('href')
    result.append(url)
    print(url)
  
  num = num + 50
  
result
```

```python
result[4]
```

```python
product_detail = {}
for x in range(len(result)):
  url = result[x]
  res = requests.get(url)
  data = BeautifulSoup(res.text)

  # get product information
  product_name = data.select('h1.ui-pdp-title')[0].string
  price = data.select('meta[itemprop=price]')[0].get('content')
  images = data.select('img.ui-pdp-image.ui-pdp-gallery__figure__image')
  images = [l.get('data-zoom') for l in images]
  images = ', '.join(images)
  if images == '':
    images = 'No brands'
  images

  # compile product detail
  product_detail[x] = {
      "url": url,
      "product_name": product_name,
      "price": price,
      "images": images
  }

  print(product_detail[x])
  
```

```python
headers = [ "url", "product_name", "price", "images"]
df = pd.DataFrame.from_dict(product_detail, orient='index')
df.to_excel("product_detail.xlsx")
files.download('product_detail.xlsx')
```
